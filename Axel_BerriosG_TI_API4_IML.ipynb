```python
from sklearn import datasets

iris= datasets.load_iris() #Este código te da una descripción de la base de datos de la flor del iris, incluyendo detalles sobre las características y las clases objetivo.

print(iris.DESCR)
```

```python
import matplotlib.pyplot as plt
```

```python
sepal_length = iris.data[:,0]
sepal_width = iris.data[:,1]

plt.scatter(sepal_length, sepal_width)
plt.xlabel('Largo del sepalo')
plt.ylabel('Ancho del sepalo')
plt.show()
```

Este código muestra una gráfica de dispersión de la longitud del sépalo en el eje x y el ancho del sépalo en el eje y, lo cual te permite visualizar las relaciones entre estas características.

```python
from sklearn.cluster import KMeans #Entrenar el modelo y determinar centros, inercia y etiqueta:
```

```python
k = 3 #Número de cluster

model = KMeans (n_clusters=k)
model.fit(iris.data)

centers = model.cluster_centers_
inertia = model.inertia_
labels = model.labels_

print('Centros:', centers)
print('Inercia:', inertia)
print('Etiquetas:', labels)
```

Este código utiliza el algoritmo K-means para entrenar un modelo con un número de clusters determinado. Los centros de cada cluster, la inercia (suma de las distancias cuadradas entre muestras y sus centroides más cercanos) y las etiquetas asignadas a cada muestra se imprimen en la consola.

```python
plt.scatter(iris.data[:,0],iris.data[:,1], c=labels)
plt.scatter(centers[:,0], centers[:,1], marker='X', color = 'red')
plt.xlabel('Largo del sépalo')
plt.ylabel('Ancho del sépalo')
plt.show()
```

Este código muestra una gráfica de dispersión de las muestras de la base de datos de la flor del iris, donde cada muestra está coloreada según la etiqueta de su cluster. Los centros de cada cluster se muestran con una “x” en color rojo.


```python
inertias = []
k_range = range(1,10)

for k in k_range:
  model = KMeans(n_clusters=k)
  model.fit(iris.data)
  inertias.append(model.inertia_)

plt.plot(k_range, inertias)
plt.xlabel('Número de clusters')
plt.ylabel('Inercia')
plt.show()
```

Este código itera sobre diferentes valores de k (número de clusters) y calcula la inercia para cada uno. Luego, muestra una gráfica de la inercia en función del número de clusters. Puedes visualizar el “codo” en la gráfica, que indica el número de clusters ideales donde el rendimiento mejora menos significativamente a medida que aumenta k.
